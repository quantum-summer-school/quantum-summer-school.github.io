# Introduction to SCQubits

```python
import scqubits as scq
import sccircuitbuilder as sc
import numpy as np
```

# LC Harmonic Oscillator

Circuit Design

```python
sc.GUI()
```

```python
H_Osc = scq.Circuit("hosc.yml.yml", from_file=True, ext_basis="harmonic")
```

Print Hamiltonian

```python
H_Osc.sym_hamiltonian()
```

```python
H_Osc.eigenvals()
```

Plot Potential

```python
H_Osc.plot_potential(θ1=np.linspace(-np.pi, np.pi))
```

Plot Wavefunction

```python
H_Osc.cutoff_names
```

```python
H_Osc.cutoff_ext_1 = 200
H_Osc.plot_wavefunction(which=2, mode="real")
```

```python
tmon_2 = scq.Circuit("transmons_2.yml.yml.yml", from_file=True)
```

```python
tmon2_yaml = """# Transmons Coupled
branches:
- [JJ, 1,2, 30, 0.01]
- [C, 1,2, EC=0.02]
- [C, 1,3, EC=0.02]
- [JJ, 3,4, 30, 0.01]
- [C, 3,4, EC=0.02]
"""
```

```python
tmon_2.sym_hamiltonian()
```

```python
tmon_2.eigenvals()
```

```python
tmon_2.plot_potential(θ1=np.linspace(-np.pi, np.pi), θ2=0)
```

```python
tmon_2.cutoff_names
```

```python
tmon_2.cutoff_n_1 = 200
tmon_2.plot_wavefunction(which=0, var_indices=(1,2), mode="real", change_discrete_charge_to_phi = True)
```
